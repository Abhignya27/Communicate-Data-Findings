## Preliminary Wrangling

> The dataset is about LA from Jan 2019 to Dec 2019.
The dataset consists of four La csv files which are divided into 4 quaters. **station.csv** consists of station name in LA.
The columns in LA dataset are -
1. **trip-id**- unique id of the trip
2. **Duration**- duration of the trip in minutes
3. **start_time**- start date and time of the trip
4. **start_station**- start station id of the trip
5. **start_lat**- the latitude of the start station
6. **start_lon**- the longitude of the start station
7. **end_station**- end station id 
8. **end_lat**- the latitude of the end station
9. **end_lon**- the longitude of the end station
10. **bike_id**- Locally unique integer that identifies the bike
11. **plan_duration**- The number of days that the plan the passholder is using entitles them to ride; 0 is used for a single       ride plan (Walk-up) 
12. **trip_route_category**-  "Round Trip" for trips starting and ending at the same station or "One Way" for all other trips
13. **passholder_type**- the passes which user holds
14. **bike_type**- The kind of bike used on the trip, including standard pedal-powered bikes, electric assist bikes, or smart      bikes.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import calendar
import plotly as px
import math
%matplotlib inline
```

```python
la1=pd.read_csv("LA-q1.csv")
la2=pd.read_csv("LA-q2.csv")
la3=pd.read_csv("LA-q3.csv")
la4=pd.read_csv("LA-q4.csv")
station=pd.read_csv("stations.csv")
```

```python
data=pd.concat([la1,la2,la3,la4])
```

```python
data[:3]
```

```python
station[:5]
```

```python
data=data.rename(columns={'start_station':'Station_ID'})
```

```python
#merging the la and station dataset
data=data.merge(station,on='Station_ID')
```

```python
data[:5]
```

```python
# droping the start_lat, start_lon, end_lat and end_lon columns
data.drop(['start_lat','start_lon','end_lat','end_lon','Region '], axis=1, inplace= True)
```

```python
# Print the details of LA
print(data.shape)
print("")
print(data.dtypes)
print(data.head(5))
```

```python
#making a separate column for year,month,day
data['start_time'] = pd.to_datetime(data['start_time'])
data['start_year']= data['start_time'].dt.year
data['start_month']=data['start_time'].dt.month
data['start_day']=data['start_time'].dt.day
data[:3]
```

```python
data['start_hour'] = data['start_time'].apply(lambda time: time.hour)
data['period_day'] = 'morning'
data['period_day'][(data['start_hour'] >= 12) & (data['start_hour'] <= 17)] = 'afternoon'
data['period_day'][(data['start_hour'] >= 18) & (data['start_hour'] <= 23)] = 'night'
```

```python
data.period_day.value_counts()
```

```python
data['seasons']='Winter'
data['seasons'][(data['start_month'] >= 3) & (data['start_month'] <= 5)]='Spring'
data['seasons'][(data['start_month'] >=6) & (data['start_month'] <= 8)]='Summer'
data['seasons'][(data['start_month'] >=9) & (data['start_month'] <= 11)]='Autumn'
```

```python
data.seasons.value_counts()
```

```python
data['week_num'] = data['start_time'].apply(lambda time: time.dayofweek)
d = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
data['week_name'] = data['week_num'].map(d)
```

```python
data.week_num.value_counts()
```

```python
data.week_name.value_counts()
```

```python
# converting the month number to month name
data['start_month'] = data['start_month'].apply(lambda x: calendar.month_abbr[x])
```

```python
data['Station_Name'].value_counts()
```

```python
ap = data.query('passholder_type == "Annual Pass"')
ap_counts= ap.groupby(['week_name','start_month']).size()
ap_counts= ap_counts.reset_index(name='count')
ap_counts = ap_counts.pivot(index='week_name', columns='start_month', values='count')
```

```python
fp = data.query('passholder_type == "Flex Pass"')
fp_counts= fp.groupby(['week_name','start_month']).size()
fp_counts= fp_counts.reset_index(name='count')
fp_counts = fp_counts.pivot(index='week_name', columns='start_month', values='count')
```

```python
mp = data.query('passholder_type == "Monthly Pass"')
mp_counts= mp.groupby(['week_name','start_month']).size()
mp_counts= mp_counts.reset_index(name='count')
mp_counts = mp_counts.pivot(index='week_name', columns='start_month', values='count')
```

```python
op = data.query('passholder_type == "One Day Pass"')
op_counts= op.groupby(['week_name','start_month']).size()
op_counts= op_counts.reset_index(name='count')
op_counts = op_counts.pivot(index='week_name', columns='start_month', values='count')
```

```python
test = data.query('passholder_type == "Testing"')
test_counts= test.groupby(['week_name','start_month']).size()
test_counts= test_counts.reset_index(name='count')
test_counts = test_counts.pivot(index='week_name', columns='start_month', values='count')
```

```python
walk = data.query('passholder_type == "Walk-up"')
walk_counts= walk.groupby(['week_name','start_month']).size()
walk_counts= walk_counts.reset_index(name='count')
walk_counts = walk_counts.pivot(index='week_name', columns='start_month', values='count')
```
